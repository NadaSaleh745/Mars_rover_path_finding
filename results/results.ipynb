```python
import matplotlib.pyplot as plt
import pandas as pd

results = pd.read_csv('/Users/nada/PycharmProjects/Mars_rover_path_finding /Mars_rover_path_finding/results/search_algorithm_results.csv')
```

```python
results
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

axes[0].plot(results["Algorithm"], results["Path Length"], marker='o')
axes[0].set_ylabel("Path Length")
axes[0].set_title("Path Length by Algorithm")
axes[0].tick_params(axis='x', rotation=45)

axes[1].plot(results["Algorithm"], results["Nodes Explored"], marker='o')
axes[1].set_ylabel("Nodes Explored")
axes[1].set_title("Nodes Explored by Algorithm")
axes[1].set_yscale("log")
axes[1].tick_params(axis='x', rotation=45)

axes[2].plot(results["Algorithm"], results["Time (ms)"], marker='o')
axes[2].set_ylabel("Time (ms)")
axes[2].set_title("Execution Time by Algorithm")
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
